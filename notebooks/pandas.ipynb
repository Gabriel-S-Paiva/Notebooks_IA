### Step 3: Cleaning

Some ML models (like Scikit-Learn) dont accept NaN

```python
print('Null values per collumn', df.isnull().sum())
```

```python
df['PetalLengthCm'] = df['PetalLengthCm'].fillna(df['PetalLengthCm'].mean())
df['PetalWidthCm'] = df['PetalWidthCm'].fillna(df['PetalWidthCm'].mean())

print('\nNot null after treating', df.isnull().sum().sum())
```

### Step 4: Feature engineering

ML algorithms work using math, so they need numbers. Lets work on the Species collumn

```python
species_map = {
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
}

df['Species_numeric'] = df['Species'].map(species_map)

print('Trasformation aplied')
print(df[['Species', 'Species_numeric']].sample(10))
```

### Step 5: Correlation Anylasis

What features are more important to predict the species

```python
df_num = df.drop('Species', axis=1)

print('Correlation Matrix')
print(df_num.corr())
```
